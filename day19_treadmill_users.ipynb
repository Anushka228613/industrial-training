```python
#Categorical Data
for col in cat_cols:
    plt.figure(figsize=(10,3))
    plt.subplot(1,2,1)
    sns.countplot(data=data,x=col,palette=choice(p_list))
    plt.title(col)
    plt.subplot(1,2,2)
    data[col].value_counts().plot.pie(autopct = '%1.1f%%',figsize=(8,4))
    plt.show()

#Numerical Data
for col in num_cols:
    color =[random(),random(),random()]
    fig = plt.figure(figsize=(10,3))
    fig.suptitle(col)
    plt.subplot(1,3,1)
    sns.distplot(data[col],color = color)
    plt.subplot(1,3,2)
    sns.boxplot(data = data,x = col, color =  color)
    plt.subplot(1,3,3)
    sns.violinplot(data = data,x = col, color = color)
    plt.show()
```

```python
for col in cat_cols:
    for ncol in num_cols:
        plt.figure(figsize=(10,3))
        plt.suptitle(f'{col} vs {ncol}')
        plt.subplot(1,3,1)
        sns.boxplot(data = data,x = col,y=ncol, palette=choice(p_list))
        plt.subplot(1,3,2)
        sns.violinplot(data = data, x=col,y=ncol, palette=choice(p_list))
        plt.subplot(1,3,3)
        sns.barplot(data= data, x=col,y=ncol, palette=choice(p_list))
        plt.show()

```

```python
pg = sns.PairGrid(data = data[num_cols])
pg.map_diag(plt.hist)
pg.map_lower(sns.regplot)
pg.map_upper(sns.kdeplot)
```

```python
sns.heatmap(
    data = data[num_cols].corr(),
    annot=True,
    cmap = choice(p_list)
)

```

Multivariate Analysis

```python
for col in cat_cols:
    if col == 'Product':
        continue
    
    for ncol in num_cols:
        plt.figure(figsize=(10,3))
        plt.suptitle(f'{col} vs {ncol}')
        plt.subplot(1,3,1)
        sns.boxplot(data = data,x = col,y=ncol,hue='Product', palette=choice(p_list))
        plt.subplot(1,3,2)
        sns.violinplot(data = data, x=col,y=ncol,hue='Product',palette=choice(p_list))
        plt.subplot(1,3,3)
        sns.barplot(data= data, x=col,y=ncol,hue='Product', palette=choice(p_list))
        plt.show()
```
