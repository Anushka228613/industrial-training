```python
import matplotlib.pyplot as pit
import numpy as np
```

```python
#Types of Graphs
#1.plot-         line graph,between two parameters 
#2.bar -         bar graph of a two parameter  
#3.hist -        histogram of a bunch of a dat i.e. frequency of two parameter
#4.scatter -     dot graph between two parameters  
#5.boxplot -     statistical overview in form of box for a single parameter   mean,outliers,quantiles(25%,75%),IQR or categorical data
#6.stem  -       more like bar graoh ,lines instead of bars
#7.violinplot -  violin shaped statistical overview graph 
#                spread of data(more like histogram)
#                single parameter (optional cat - data)
```

```python
x= np.linspace(0,5,11)
x1=np.arange(1,11,3)
```

```python
x
```

```python
x1
```

```python
y = x**3
```

```python
pit.plot(x,y)
#pit.show()
```

```python
pit.bar(x,y)
pit.title('Bar Graph')
```

```python
pit.hist(x,y)
pit.title('Hist Graph')
```

```python
pit.boxplot(x)
pit.title('Boxplot Graph')
```

```python
pit.stem(x,y)
pit.title('Stem Graph')
```

```python
pit.violinplot(x)
pit.title('Violinplot Graph')
```

```python
pit.scatter(x,y)
pit.title('Scatter Graph')
pit.grid()
pit.xlabel('x-->')
pit.ylabel('y-->')
pit.xticks(x)
pit.yticks(range(0,200,10))
pit.show()
```

```python
y2=x**4
```

```python
pit.plot(x,y)
pit.show()
pit.plot(x,y2)
pit.show()
```

```python
pit.plot(x,y,x,y2)
pit.legend(['y=x**3','y2=x**4'])
pit.show()
```

```python
fig =pit.figure(figsize=(10,5))
pit.plot(x,y,x,lw=3,marker='o',markersize=15,markerfacecolor='yellow')
pit.plot(x,y,marker='v',markersize=10)
pit.legend(['y=x**3','y2=x**4'])
pit.show()
```
